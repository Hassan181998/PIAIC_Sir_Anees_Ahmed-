```python
# Importing Libraries
import numpy as np
from PIL import Image
import os
import glob
```

```python
#  Function which loads image into into an array from folder photos
def loading_image():
    images = []
    try:
        for pictures in glob.glob(r'photos\*.jpg'):
            images.append(Image.open(pictures))
    except IOError:
        print('Invalid File Directory.')
    return images

# below function resizes the image to (200,200)         
def resizing_image():
    resized_images =[]
    loaded_images = loading_image()
    for image in loaded_images:
        resized_images.append(image.resize([200,200]))
    return resized_images

# the resized images are converted into numpy array
def convert_into_numpy_array():
    temp_arr = resizing_image()
    final_img_arr = []
    for items in temp_arr:
        numpy = np.array(items)
        final_img_arr.append(numpy)
#         print(np.shape(final_img_arr),end = "\n")
#         print(final_img_arr,'\n',np.shape(final_img_arr))
        
    return numpy,final_img_arr
    
# The main method print numpy array of 20 images
def main():
    while True :
        print('\t\t\t\t\t *********** CONVERTING IMAGES TO NUMPY ARRAY ************\t\t\t\t\t\t\n',
             'Press 1: To see Numpy Array of each resized images.\n',
             'Press 2: To see combined numpy array of all 20 images.\n')

        # Calling function which converted images to numpy array it returns  2 things 1. combined np array 2. single np array
        single_img_arr, all_images_arr =convert_into_numpy_array()
        user_input = int(input())
        if user_input == 1:
#         new_resized_arr = resizing_image()
            pic_no = 1
            print('Below are the numpy array of each image after being resized :\n')
#         print(len(single_img_arr))
            for image in single_img_arr :
                print("Picture ", pic_no, image, end ='\n')
                pic_no += 1
            break
        elif user_input == 2:
            print("Below is the numpy array which holds all resized array of 20 images in a single numpy array :\n")
            print(all_images_arr)
            break
        else:
            print('You have entered wrong key. Please try again with correct options.')
            continue

        
if __name__ =="__main__":
    main()
    
```
